```python
from sklearn.model_selection import cross_val_score,train_test_split,cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score,average_precision_score,precision_recall_curve
import itertools
```

```python
#all attributes excluding target variable ,asthma,and MI_6mo
X = df1.drop(['Death_1yr','MI_6mo','Asthma'],axis=1)
y = df1['Death_1yr']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=1111,stratify=y)
```

```python
def plot_confusion_matrix(cm, classes, title='Confusion matrix', cmap=plt.cm.Blues):

    print('Confusion matrix')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt ='d'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
def model_report(model,title):

    clf = model(random_state=1111)

    clf.fit(X, y)


    class_names = ['Live', 'Death']

    num_folds = 10
    seed = 1111
    kfold = KFold(n_splits=num_folds, random_state=seed,shuffle=True)
    results = cross_val_score(clf,X,y, cv=kfold,scoring='accuracy')

    precision=cross_val_score(clf,X,y,cv=kfold,scoring='precision_weighted')
    recall=cross_val_score(clf,X,y,cv=kfold,scoring='recall_weighted')
    print('Accuracy in Each Fold')
    print(results)
    print('----------------------------------------------------')
    print('Precision in Each Fold')
    print(precision)
    print('----------------------------------------------------')
    print('Recall in Each Fold')
    print(recall)
    print('----------------------------------------------------')

    pre1=precision.mean();
    rec1=recall.mean()
    print("Mean Accuracy:", results.mean()*100.0)
    print("Mean precision:", precision.mean()*100.0)
    print("Mean recall:", recall.mean()*100.0)
    print("STD Accuracy:", results.std()*100.0)
    fscore=2*((pre1*rec1)/(pre1+rec1))
    print("F score",fscore)

    print('-----------------------------------------------------------------------------')

    y_pred_kfold=clf.predict(X_test)





    print("Confusion matrix with k folds")


    print('Average Precision: {:.2f}'.format(average_precision_score(y_test, y_pred_kfold)))
    print(classification_report(y_test, y_pred_kfold, target_names=class_names))

    cnf_matrix = confusion_matrix(y_test, y_pred_kfold)
    np.set_printoptions(precision=2)

    # Plot confusion matrix
    plt.figure()
    plot_confusion_matrix(cnf_matrix, classes=class_names)
    plt.show()

    #precision-recall curve
    from inspect import signature
    import pylab as pl
    print('------------------------------------------------------------')
    average_precision=average_precision_score(y_test,y_pred_kfold)

    prec, rec, tre = precision_recall_curve( y_test, y_pred_kfold)
#pl.clf()
    pl.plot(rec, prec, label='Precision-Recall curve')
    pl.xlabel('Recall')
    pl.ylabel('Precision')
    pl.ylim([0.0, 1.0])
    pl.xlim([0.0, 1.0])
    pl.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(average_precision))
    pl.legend(loc="lower left")
    pl.show()

    print('-----------------------------------------------------------')




    #prec, rec, tre = precision_recall_curve( y_test, y_score_rf)
    step_kwargs = ({'step': 'post'} if 'step' in signature(plt.fill_between).parameters else {})
    plt.step(rec, prec, color='r', alpha=0.2, where='post')
    plt.fill_between(rec, prec, alpha=0.2, color='r', **step_kwargs)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.0])
    plt.xlim([0.0, 1.0])
    plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(average_precision))


```
